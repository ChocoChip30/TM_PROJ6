## Preparing the Data

```python
df['Contents'] = (df['Allergens'] + ' ' + df['Ingredients']).str.lower()
df.head(3)
```

```python
import re

def extract_clean_words(text):
    words = re.findall(r'\b[a-zA-Z]+\b', text)  
    cleaned_words = [word for word in words if word.isalpha()]  
    return cleaned_words

df['Cleaned_Words'] = df['Contents'].apply(extract_clean_words)
```

```python
df.drop(columns=['Allergens', 'Ingredients', 'Contents'], inplace=True)
df.head(3)
```

#### Creating separate columns to indicate the presence of certain ingredients

```python
def set_milk(row):
    if 'milk' in row['Cleaned_Words']:
        return 1
    else:
        return 0
    
def set_egg(row):
    if 'egg' in row['Cleaned_Words']:
        return 1
    if 'eggs' in row['Cleaned_Words']:
        return 1
    else:
        return 0
    
def set_nuts(row):
    if 'nuts' in row['Cleaned_Words']:
        return 1
    elif 'nut' in row['Cleaned_Words']:
        return 1
    elif 'almonds' in row['Cleaned_Words']:
        return 1
    elif 'cashews' in row['Cleaned_Words']:
        return 1
    elif 'hazlenuts' in row['Cleaned_Words']:
        return 1
    elif 'peanuts' in row['Cleaned_Words']:
        return 1
    else:
        return 0
    
def set_soy(row):
    if 'soy' in row['Cleaned_Words']:
        return 1
    elif 'soya' in row['Cleaned_Words']:
        return 1
    else:
        return 0
    
df['Milk'] = df.apply(set_milk, axis=1)
df['Egg'] = df.apply(set_egg, axis=1)
df['Nuts'] = df.apply(set_nuts, axis=1)
df['Soy'] = df.apply(set_soy, axis=1)
```

```python
df.tail(3)
```

```python
df2 = df[['Name', 'Calories/g', 'Protein/g', 'Carbohydrate/g', 'Sugar/g',
       'Fat/g', 'Fibre/g', 'Protein/kcal','Price/g', 'Vegan',
       'GlutenFree', 'Flavor','Milk', 'Egg', 'Nuts', 'Soy']]
df2.tail(3)
```

#### One hot encoding on 'Flavor' Column

```python
df2.Flavor = df2['Flavor'].str.lower()
```

```python
dummies = pd.get_dummies(df2.Flavor,  dtype='int')
dummies.head(3)
```

```python
#dropping last col to avoid trap
df3 = pd.concat([df2, dummies.drop('caramel',axis='columns' )], axis='columns')

df3.head(3)
```

```python
df4 = df3.copy()
```

```python
df4.drop(columns=['Name'], inplace=True)
df4.drop(columns=['Flavor'], inplace=True)
```

```python
df4.info()
```

#### Min-Max Scaling on DataFrame

*Min-max scaling is a technique used in machine learning and data preprocessing to scale the features of a dataset so that they fall within a specific range, typically between 0 and 1. This is achieved by subtracting the minimum value of each feature and then dividing by the difference between the maximum and minimum values of that feature.*

```python
df5 = (df4 - df4.min()) / (df4.max() - df4.min())
df5.head()
```

```python
df6 = df5.copy()
df6['Name'] = df3['Name']
df6.head(3)
```

## Finalising the Data

```python
last_column = df6.iloc[:, -1]

df6 = df6.drop(df6.columns[-1], axis=1)

df6.insert(0, last_column.name, last_column)

df6.head()
```

```python
df6.insert(0, 'ID', range(1, len(df) + 1))
df6.head()
```
